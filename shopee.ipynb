```python
import requests
import json
import time
from bs4 import BeautifulSoup
from urllib import parse

# 正則表達式用
import re
```

```python
headers = {'User-Agent': 'Googlebot',}
def sleeptime(hour,min,sec):
    return hour*3600 + min*60 + sec;
```

```python
def getProduct(product_id):
    products = []
    try:
        # get product information
        for item_id, shop_id in product_id:
            url = 'https://shopee.tw/api/v2/item/get?itemid={}&shopid={}'.format(item_id, shop_id)
            response = requests.get(url, headers=headers, allow_redirects=False)
            item = json.loads(response.text)['item']
            products.append([item['price_max']//100000,
                             item['item_rating']['rating_star'],
                             item['item_rating']['rating_count'][0],
                             item['liked_count'],
                             item['historical_sold'],
                            ])
    except:
        print("getProduct excpetion")
    return products
```

```python
def getProductID(keyword):
    product_id = []
    try:
        # get itemid, shopid
        for index in range(150):
            url = 'https://shopee.tw/api/v2/search_items/?by=relevancy&keyword={}&limit=50\
                    &newest={}&order=desc&page_type=search&version=2'.format(keyword, index * 50)
            response = requests.get(url, headers=headers, allow_redirects=False)
            data = json.loads(response.text)
            for item in data['items']:
                product_id.append([item['itemid'], item['shopid']])
    except:
        print('getProductID exception', index)
    finally:
        return product_id
```

```python
product_id = getProductID('藍芽耳機')
```

```python
print(product_id, len(product_id))
```

```python
import time
try:
    file = open('shopee商品資料.txt', 'a')
    for index in range(0, len(product_id), 100):
        products = getProduct(product_id[index:index+100])
        for product in products:
            print(*product, sep=',', file=file)
        time.sleep(sleeptime(0,1,0))
    file.close()
except:
    print('except', index)
```
